### **Data Loading and Cleansing - Part 2**

### Loading

Next, I will work on loading the 2016 data.

- The 2026 data is in single csv file per produce and have been saved to the folder Fruits_Vegs_2016.
- The name of the files are in following format: "<produce name> 2016.csv"; for example: "apples 2016.csv"
- Only the first 7 columns contain information. 
- Some files have additional columns with null values.

- An empty dataframe (df_produces_2016) will be created.
- For each file in the Fruits_Vegs_2016 folder, the following will be done:
    - Get the filename.
    - Extract the produce name from the file name.
    - Read the file into a dataframe.
    - If the dataframe have more than 7 columns, additional columns will be dropped.
    - The columns will be renamed to be consistent with the naming of the columns in df_produces (containing 2020 and 2022 data)
    - A column with the produce name will be added to the dataframe.
    - Concatenate the dataframe to dataframe df_produces_2026


### The following a sample 2026 file.

<img src="2016_sample_file.png" />


```python
# Assign directory
directory = r"Fruits_Vegs_2016\\" 
produces_2016  = pd.DataFrame()

# Iterate over files in directory
for fname in os.listdir(directory):
    pos = fname.find(" ")
    produce_nm = fname[:pos]
    complete_filename = directory + fname
    produce = pd.read_csv(complete_filename, encoding='ISO-8859-1')

    #Information in these files are stored in the first 7 columns. The following drops any additional column.
    column_cnt = produce.shape[1]
    if column_cnt > 7:
        produce = produce.drop(produce.iloc[:,7:column_cnt], axis=1)   
    
    #The first row contains a description of the file, wwhich is read as the first column heading and the others are unnamed.
    #Therefore, the columns are renamed to be consistent with the columns in the dataframe with the 20202 and 2022 data.
   
    produce = produce.set_axis(['Form', 'RetailPrice', 'RetailPriceUnit', 'Yield', 'CupEquivalentSize', 'CupEquivalentUnit', 'CupEquivalentPrice'], axis=1)
    #df_produce['Produce'] = produce_nm
    produce.insert(0, "Produce", produce_nm)
    produces_2016 = pd.concat([produces_2016,produce], axis=0).reset_index(drop=True)
 
produces_2016.head(10)
```

Create dataframe with all_combinations as follows:
1. Combine records with the first 2 colors.
2. Drop combinations that includes records from multiple years
3. For each additional color, repeat steps 1 & 2 by combining prior result with that color

Create fresh_combinations by extracting records from all combinations were all form column equal fresh.
Create canned_combinations by extracting records from all combinations were all form column equal canned.
Create frozen_combinations by extracting records from all combinations were all form column equal frozen.

```python
v_colors = classified_produce['Color'].drop_duplicates()
red = classified_produce[classified_produce['Color'] == 'Red']
green = classified_produce[classified_produce['Color'] == 'Greens']
yellow = classified_produce[classified_produce['Color'] == 'Yellow/Orange']
white = classified_produce[classified_produce['Color'] == 'White/Tan/Brown']
blue = classified_produce[classified_produce['Color'] == 'Blue/Purple']

#renaming columns to support the cartisian merge
red = red.set_axis(['Color_r', 'Produce_r', 'Form_r', 'CupEquivalentPrice_r', 'Year_r'], axis=1)
green = green.set_axis(['Color_g', 'Produce_g', 'Form_g', 'CupEquivalentPrice_g', 'Year_g'], axis=1)
yellow = yellow.set_axis(['Color_y', 'Produce_y', 'Form_y', 'CupEquivalentPrice_y', 'Year_y'], axis=1)
white = white.set_axis(['Color_w', 'Produce_w', 'Form_w', 'CupEquivalentPrice_w', 'Year_w'], axis=1)
blue = blue.set_axis(['Color_b', 'Produce_b', 'Form_b', 'CupEquivalentPrice_b', 'Year_b'], axis=1)

all_combinations = blue.assign(key=1).merge(white.assign(key=1), on='key').drop('key', axis=1)
all_combinations = all_combinations[all_combinations['Year_b'] == all_combinations['Year_w']]
all_combinations = all_combinations.assign(key=1).merge(yellow.assign(key=1), on='key').drop('key', axis=1)
all_combinations = all_combinations[all_combinations['Year_b'] == all_combinations['Year_y']]
all_combinations = all_combinations.assign(key=1).merge(green.assign(key=1), on='key').drop('key', axis=1)
all_combinations = all_combinations[all_combinations['Year_b'] == all_combinations['Year_g']]
all_combinations  = all_combinations.assign(key=1).merge(red.assign(key=1), on='key').drop('key', axis=1)
all_combinations = all_combinations[all_combinations['Year_b'] == all_combinations['Year_r']]

all_combinations['Total_Price'] = all_combinations['CupEquivalentPrice_r'] + all_combinations['CupEquivalentPrice_g'] + all_combinations['CupEquivalentPrice_y'] + all_combinations['CupEquivalentPrice_w'] + all_combinations['CupEquivalentPrice_b']
all_combinations.drop(columns=['Year_w','Year_y', 'Year_g','Year_r'], inplace=True)
all_combination = all_combinations.rename(columns={'Year_b': 'Year'})

print(all_combinations.shape)
all_combinations.head(10)
```
