### **Data Loading and Cleansing - Part 2**

### Loading

Next, I will work on loading the 2016 data.

- The 2026 data is in single csv file per produce and have been saved to the folder Fruits_Vegs_2016.
- The name of the files are in following format: "<produce name> 2016.csv"; for example: "apples 2016.csv"
- Only the first 7 columns contain information. 
- Some files have additional columns with null values.

- An empty dataframe (df_produces_2016) will be created.
- For each file in the Fruits_Vegs_2016 folder, the following will be done:
    - Get the filename.
    - Extract the produce name from the file name.
    - Read the file into a dataframe.
    - If the dataframe have more than 7 columns, additional columns will be dropped.
    - The columns will be renamed to be consistent with the naming of the columns in df_produces (containing 2020 and 2022 data)
    - A column with the produce name will be added to the dataframe.
    - Concatenate the dataframe to dataframe df_produces_2026


### The following a sample 2026 file.

<img src="2016_sample_file.png" />


```python
# Assign directory
directory = r"Fruits_Vegs_2016\\" 
df_produces_2016  = pd.DataFrame()

# Iterate over files in directory
for fname in os.listdir(directory):
    pos = fname.find(" ")
    produce_nm = fname[:pos]
    complete_filename = directory + fname
    df_produce = pd.read_csv(complete_filename, encoding='ISO-8859-1')

    #Information in these files are stored in the first 7 columns. The following drops any additional column.
    column_cnt = df_produce.shape[1]
    if column_cnt > 7:
        df_produce = df_produce.drop(df_produce.iloc[:,7:column_cnt], axis=1)   
    
    #The first row contains a description of the file, wwhich is read as the first column heading and the others are unnamed.
    #Therefore, the columns are renamed to be consistent with the columns in the dataframe with the 20202 and 2022 data.
   
    df_produce = df_produce.set_axis(['Form', 'RetailPrice', 'RetailPriceUnit', 'Yield', 'CupEquivalentSize', 'CupEquivalentUnit', 'CupEquivalentPrice'], axis=1)
    #df_produce['Produce'] = produce_nm
    df_produce.insert(0, "Produce", produce_nm)
    df_produces_2016 = pd.concat([df_produces_2016,df_produce])


 
df_produces_2016.head(10)
```
