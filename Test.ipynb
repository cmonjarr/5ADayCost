```python
# Assign directory
directory = r"Fruits_Vegs_2016\\" 
df_produces_2016  = pd.DataFrame()

# Iterate over files in directory
for fname in os.listdir(directory):
    pos = fname.find(" ")
    produce_nm = fname[:pos]
    complete_filename = directory + fname
    df_produce = pd.read_csv(complete_filename, encoding='ISO-8859-1')

    #Information in these files are stored in the first 8 columns. The following drops any additional column.
    column_cnt = df_produce.shape[1]
    if column_cnt > 7:
        df_produce = df_produce.drop(df_produce.iloc[:,7:column_cnt], axis=1)   
    
    #The first row contains a description of the file, wwhich is read as the first column heading and the others are unnamed.
    #Therefore, the columns are renamed to be consistent with the columns in the files for 20202 and 2022
    #This also enable merging them into a single dataframe
    df_produce = df_produce.set_axis(['Form', 'RetailPrice', 'RetailPriceUnit', 'Yield', 'CupEquivalentSize', 'CupEquivalentUnit', 'CupEquivalentPrice'], axis=1)
    df_produce['produce_name'] = produce_nm
    df_produces_2016 = pd.concat([df_produces_2016,df_produce])


 
df_produces_2016.head(10)
```

```python
# Now, I am going to drop rows that do not containd the data needed for the analysis: 
#   Rows were the RetailPriceUnit field is null 
df_produces_2016.dropna(subset=['RetailPriceUnit'], inplace=True) 
```
