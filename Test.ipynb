```python
# Assign directory
directory = r"Fruits_Vegs_2016\\" 
df_produces_2016  = pd.DataFrame()
file_count = 0
# Iterate over files in directory
for fname in os.listdir(directory):
    pos = fname.find(" ")
    produce_nm = fname[:pos]
    complete_filename = directory + fname
    df_produce = pd.read_csv(complete_filename, encoding='ISO-8859-1')
    #print(fname)

    #Some files have some empty columns. If the file had multiple null columns, dropna was only propping 1 column; therfore, I came up with the following logic.
    #df_produce.dropna(how='all', axis=1, inplace=True)
    column_cnt = df_produce.shape[1]
    if column_cnt > 7:
        df_produce = df_produce.drop(df_produce.iloc[:,7:column_cnt], axis=1)
    

    
    
    # columns_nm = list(df_produce.columns)
    # print(columns_nm)
    # print(len(columns_nm))

    # if len(columns_nm) > 7:
    #     pos = 7
    # else: 
    #     pos = len(columns_nm)

    # while pos < len(columns_nm):
    #     #column_nm = 'Unnamed: '+str(pos)
    #     #df_produce = df_produce.drop(column_nm)
    #     df_produce = df_produce.drop(columns_nm[pos])
    #     pos += 1
    # #Finish dropping empty columns
    
    df_produce.dropna(how='all')
    df_produce = df_produce.set_axis(['Form', 'RetailPrice', 'RetailPriceUnit', 'Yield', 'CupEquivalentSize', 'CupEquivalentUnit', 'CupEquivalentPrice'], axis=1)
    df_produce['produce_name'] = produce_nm
    df_produces_2016 = pd.concat([df_produces_2016,df_produce])

# Drop records that are not needed.
df_produces_2016.dropna(subset=['RetailPrice']) 
 
df_produces_2016.value_counts("Form")
```
